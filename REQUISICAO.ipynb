```python
import websocket
import rel
import json
import pandas
import numpy
from datetime import datetime, timedelta

os=[]
cs=[]
hs=[]
ls=[]
vs=[]
ts=[]
pnow=[]

td = timedelta(minutes = 1)

data = {"RealTime":"-","Open":os,"High":hs,"Low":ls,"Close":cs,"Volume":vs} #Cria dicion

cc =input('Digite o par de moedas:(ex : btcusdt)')
interval = input('Digite o intervalo de analise da moeda:(ex : 1m 1h 1d 1M)')


socket=f"wss://stream.binance.com:9443/ws/{cc}@kline_{interval}"
socket2=f"wss://stream.binance.com:9443/ws/{cc}@trade"

def on_message(ws, message):
    
    json_message = json.loads(message)
    candle = json_message['k']
    is_candle_closed = candle['x']
    o = candle['o']
    c = candle['c']
    h = candle['h']
    l = candle['l']
    v = candle['v']
    

    if  is_candle_closed:
        os.append(float(o))
        hs.append(float(h))
        ls.append(float(l))
        cs.append(float(c))
        vs.append(float(v))
        op = ((datetime.today()))
        tot = (op-td)
        tot = tot.strftime('%Y-%m-%d %H:%M:%S')
        ts.append(str(tot))
        table=pandas.DataFrame(data,index=ts)
        print(table)
    print(datetime.today().strftime('%Y-%m-%d %H:%M:%S')+"           {:.2f}".format(float(o))+"  {:.2f}".format(float(h))+"  {:.2f}".format(float(l))+"  {:.2f}".format(float(c))+"  {:.2f}".format(float(v)), end='\r')
        
def on_error(ws, error):
    print(error)

def on_close(ws, close_status_code, close_msg):
    print("### closed ###")

def on_open(ws):
    print("Opened connection [Candles]")
    
################################## connect 2  
def on_message2(ws2, message):
    
    json_message = json.loads(message)
    p = json_message['p']
    p = float(p)
    pnow.append(float(p))
    print(datetime.today().strftime('%Y-%m-%d %H:%M:%S')+" {:.2f}".format(p), end='\r')
    
def on_error2(ws2, error):
    print(error)

def on_close2(ws2, close_status_code, close_msg):
    print("### closed2 ###")

def on_open2(ws2):
    print("Opened connection [Price=RealTime]")
    
ws = websocket.WebSocketApp(socket,on_open=on_open,on_message=on_message,on_error=on_error,on_close=on_close)
ws.run_forever(dispatcher=rel)  # Set dispatcher to automatic reconnection
ws2 =websocket.WebSocketApp(socket2,on_open=on_open2,on_message=on_message2,on_error=on_error2,on_close=on_close2)
ws2.run_forever(dispatcher=rel)  # Set dispatcher to automatic reconnection
rel.signal(2, rel.abort)  # Keyboard Interrupt
rel.dispatch()
```
